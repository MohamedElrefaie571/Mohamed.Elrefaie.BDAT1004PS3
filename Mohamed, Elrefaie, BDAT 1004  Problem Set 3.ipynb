# Question 1

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Import dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 4: Mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean().sort_values()
print("Mean age per occupation:\n", mean_age_per_occupation)

# Visualize mean age per occupation
mean_age_per_occupation.plot(kind='bar', figsize=(10, 6), title="Mean Age per Occupation")
plt.ylabel("Mean Age")
plt.xlabel("Occupation")
plt.show()

# Step 5: Male ratio per occupation
male_ratio = (users[users['gender'] == 'M']
              .groupby('occupation')['gender']
              .count() / users.groupby('occupation')['gender'].count())
male_ratio_sorted = male_ratio.sort_values(ascending=False)
print("\nMale ratio per occupation (sorted):\n", male_ratio_sorted)

# Visualize male ratio
male_ratio_sorted.plot(kind='bar', figsize=(10, 6), title="Male Ratio per Occupation")
plt.ylabel("Male Ratio")
plt.xlabel("Occupation")
plt.show()

# Step 6: Minimum and maximum ages per occupation
min_max_age = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMinimum and Maximum Ages per Occupation:\n", min_max_age)

# Visualize minimum and maximum ages
min_max_age.plot(kind='bar', figsize=(12, 6), title="Min and Max Ages per Occupation")
plt.ylabel("Age")
plt.xlabel("Occupation")
plt.show()

# Step 7: Mean age for each occupation and gender combination
mean_age_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean Age for Each Occupation and Gender:\n", mean_age_occupation_gender)

# Visualize mean age by gender for each occupation
mean_age_occupation_gender.unstack().plot(kind='bar', figsize=(12, 6), title="Mean Age by Gender and Occupation")
plt.ylabel("Mean Age")
plt.xlabel("Occupation")
plt.xticks(rotation=45)
plt.show()

# Step 8: Percentage of men and women per occupation
gender_counts = users.groupby(['occupation', 'gender']).size()
gender_percentage = (gender_counts / gender_counts.groupby(level=0).sum() * 100).unstack()
print("\nPercentage of Men and Women per Occupation:\n", gender_percentage)

# Visualize gender percentages per occupation
gender_percentage.plot(kind='bar', stacked=True, figsize=(12, 6), title="Gender Percentage per Occupation")
plt.ylabel("Percentage")
plt.xlabel("Occupation")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.show()
```

# Question 2

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Import dataset
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

# Step 4: Select Goal column
print("Goals column:\n", euro12['Goals'])

# Visualize goals distribution
euro12['Goals'].plot(kind='hist', bins=10, figsize=(10, 6), title="Goals Distribution")
plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.show()

# Step 5: Number of teams participated
num_teams = euro12['Team'].nunique()
print(f"\nNumber of teams participated: {num_teams}")

# Step 6: Number of columns in the dataset
num_columns = euro12.shape[1]
print(f"\nNumber of columns in the dataset: {num_columns}")

# Step 7: Yellow and Red Cards
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\nYellow and Red Cards:\n", discipline)

# Step 8: Sort by Red Cards then Yellow Cards
sorted_discipline = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False])
print("\nTeams sorted by Red Cards and Yellow Cards:\n", sorted_discipline)

# Visualize Red and Yellow Cards by Team
sorted_discipline.set_index('Team').plot(kind='bar', stacked=True, figsize=(12, 6), title="Red and Yellow Cards per Team")
plt.ylabel("Number of Cards")
plt.xlabel("Team")
plt.xticks(rotation=45)
plt.show()

# Step 9: Mean Yellow Cards
mean_yellow = discipline['Yellow Cards'].mean()
print(f"\nMean Yellow Cards per Team: {mean_yellow:.2f}")

# Visualize mean Yellow Cards as a reference line
discipline['Yellow Cards'].plot(kind='box', figsize=(6, 8), title="Yellow Cards Distribution")
plt.axhline(mean_yellow, color='red', linestyle='--', label=f"Mean: {mean_yellow:.2f}")
plt.legend()
plt.show()

# Step 10: Teams scored more than 6 goals
teams_6_goals = euro12[euro12['Goals'] > 6][['Team', 'Goals']]
print("\nTeams that scored more than 6 goals:\n", teams_6_goals)

# Visualize teams with more than 6 goals
teams_6_goals.set_index('Team')['Goals'].plot(kind='bar', figsize=(10, 6), title="Teams Scored More than 6 Goals")
plt.ylabel("Goals")
plt.xlabel("Team")
plt.show()

# Step 11: Teams starting with 'G'
teams_start_g = euro12[euro12['Team'].str.startswith('G')]['Team']
print("\nTeams starting with 'G':\n", teams_start_g)

# Step 12: First 7 columns
print("\nFirst 7 columns:\n", euro12.iloc[:, :7])

# Step 13: All columns except last 3
print("\nAll columns except the last 3:\n", euro12.iloc[:, :-3])

# Step 14: Shooting Accuracy from England, Italy, and Russia
shooting_accuracy = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]

# Ensure Shooting Accuracy is numeric (remove any non-numeric characters like %)
shooting_accuracy['Shooting Accuracy'] = shooting_accuracy['Shooting Accuracy'].str.replace('%', '').astype(float)

print("\nShooting Accuracy of England, Italy, and Russia:\n", shooting_accuracy)

# Visualize shooting accuracy for England, Italy, and Russia
shooting_accuracy.set_index('Team')['Shooting Accuracy'].plot(kind='bar', figsize=(10, 6), title="Shooting Accuracy: England, Italy, Russia")
plt.ylabel("Shooting Accuracy (%)")
plt.xlabel("Team")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'yellow', 'cyan']

# Step 1 Import the necessary libraries
print("Step 1: Import necessary libraries")
# Libraries are already imported above

# Step 2 Import the dataset
print("\nStep 2: Import the dataset")
file_path = r"C:\Users\lovey\Downloads\wind.txt"
data = pd.read_csv(file_path, sep=r'\s+', header=None, na_values="NaN")
print(data.head())

# Step 3: Assign column names
print("\nStep 3: Assign column names")
columns = ['Yr', 'Mo', 'Dy', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']
data.columns = columns
print(data.head())

# Step 4: Ensure the first three columns are integers
print("\nStep 4: Convert 'Yr', 'Mo', 'Dy' columns to integers")
data[['Yr', 'Mo', 'Dy']] = data[['Yr', 'Mo', 'Dy']].apply(pd.to_numeric, errors='coerce')
print(data[['Yr', 'Mo', 'Dy']].head())

# Step 5: Create a proper datetime column
print("\nStep 5: Create a datetime column and drop invalid rows")
data['Date'] = pd.to_datetime(
    {'year': data['Yr'] + 1900, 'month': data['Mo'], 'day': data['Dy']},
    errors='coerce'
)
data = data.dropna(subset=['Date'])
print(data[['Date']].head())

# Step 6: Drop 'Yr', 'Mo', 'Dy' and set 'Date' as index
print("\nStep 6: Drop 'Yr', 'Mo', 'Dy' and set 'Date' as index")
data = data.drop(columns=['Yr', 'Mo', 'Dy'])
data.set_index('Date', inplace=True)
print(data.head())

# Step 7: Ensure all windspeed columns are numeric
print("\nStep 7: Ensure all windspeed columns are numeric")
data = data.apply(pd.to_numeric, errors='coerce')
print(data.info())

# Step 8: Compute missing values
print("\nStep 8: Compute missing values")
missing_values = data.isnull().sum()
print(missing_values)
missing_values.plot(kind='bar', figsize=(10, 6), title='Missing Values per Location', color=colors)
plt.show()

# Step 9: Compute total non-missing values
print("\nStep 9: Compute total non-missing values")
non_missing_values = data.notnull().sum().sum()
print(f"Total non-missing values: {non_missing_values}")

# Step 10: Calculate mean windspeed over all locations and times
print("\nStep 10: Calculate mean windspeed over all locations and times")
mean_windspeed = data.mean().mean()
print(f"Mean windspeed: {mean_windspeed}")

# Step 11: Calculate loc_stats
print("\nStep 11: Calculate loc_stats")
loc_stats = data.describe().T[['min', 'max', 'mean', 'std']]
print(loc_stats)
loc_stats.plot(kind='bar', figsize=(12, 6), title='Statistics per Location')
plt.show()

# Step 12: Calculate day_stats
print("\nStep 12: Calculate day_stats")
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})
print(day_stats.head())

# Prepare data for visualization: Use the first few rows to make it clearer
day_stats_subset = day_stats.head(10)  # Adjust this value for more/less rows
day_stats_subset.plot(kind='bar', figsize=(12, 6), title='Daily Wind Statistics Across Locations')

# Add labels
plt.xlabel('Date (First 10 Days)')
plt.ylabel('Windspeed (knots)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Statistics')
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()

# Step 13: Average windspeed in January
print("\nStep 13: Calculate average windspeed in January")
january_data = data[data.index.month == 1]
january_means = january_data.groupby(january_data.index.year).mean()
print(january_means)
january_means.mean().plot(kind='bar', title='Average January Windspeeds per Location', figsize=(10, 6), color=colors)
plt.show()

# Step 14: Downsample to yearly frequency
print("\nStep 14: Downsample to yearly frequency")
yearly_data = data.resample('YE').mean()
print(yearly_data.head())

# Heatmap for yearly data
print("\nHeatmap for Yearly Data")
plt.figure(figsize=(12, 8))
plt.imshow(yearly_data.T, aspect='auto', cmap="YlGnBu", interpolation='none')
plt.colorbar(label='Windspeed (knots)')
plt.title("Yearly Windspeed Heatmap")
plt.ylabel("Location")
plt.xlabel("Year")
plt.xticks(range(len(yearly_data.index)), yearly_data.index.year, rotation=45)
plt.yticks(range(len(yearly_data.columns)), yearly_data.columns)
plt.tight_layout()
plt.show()

# Step 15: Downsample to monthly frequency
import seaborn as sns  # Ensure seaborn is imported

print("\nStep 15: Downsample to monthly frequency")
monthly_data = data.resample('MS').mean()
print(monthly_data.head())

# Step 16: Downsample to weekly frequency
print("\nStep 16: Downsample to weekly frequency")
weekly_data = data.resample('W').mean()
print(weekly_data.head())

# Step 17: Calculate weekly stats (first 52 weeks)
print("\nStep 17: Calculate weekly stats (first 52 weeks)")
weekly_stats = weekly_data.iloc[:52].agg(['min', 'max', 'mean', 'std'])
print(weekly_stats)
weekly_stats.T.plot(kind='bar', figsize=(12, 6), title='Weekly Statistics (First 52 Weeks)', rot=45)
plt.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 3: Assign to a variable called chipo
# Already done above.

# Step 4: See the first 10 entries
print("First 10 entries:")
print(chipo.head(10))

# Step 5: Number of observations in the dataset
num_observations = chipo.shape[0]
print(f"\nNumber of observations: {num_observations}")

# Step 6: Number of columns in the dataset
num_columns = chipo.shape[1]
print(f"\nNumber of columns: {num_columns}")

# Step 7: Print the name of all the columns
print("\nColumn names:")
print(chipo.columns)

# Step 8: How is the dataset indexed?
print("\nDataset index:")
print(chipo.index)

# Step 9: Most-ordered item
most_ordered_item = chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(1)
print("\nMost-ordered item:")
print(most_ordered_item)

# Step 10: How many items were ordered for the most-ordered item
most_ordered_item_quantity = most_ordered_item.iloc[0]
print(f"\nQuantity of the most-ordered item: {most_ordered_item_quantity}")

# Step 11: Most ordered item in the choice_description column
most_ordered_choice = chipo.groupby('choice_description')['quantity'].sum().sort_values(ascending=False).head(1)
print("\nMost-ordered item in choice_description column:")
print(most_ordered_choice)

# Step 12: Total number of items ordered
total_items_ordered = chipo['quantity'].sum()
print(f"\nTotal items ordered: {total_items_ordered}")

# Step 13: Turn the item price into a float
# Remove the dollar sign and convert to float
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

# Check the item price type
print("\nItem price type after conversion:")
print(chipo['item_price'].dtype)

# Step 14: Calculate the total revenue for the period
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print(f"\nTotal revenue for the period: ${revenue:.2f}")

# Step 15: How many orders were made in the period
total_orders = chipo['order_id'].nunique()
print(f"\nTotal orders made: {total_orders}")

# Step 16: Average revenue per order
average_revenue_per_order = revenue / total_orders
print(f"\nAverage revenue per order: ${average_revenue_per_order:.2f}")

# Step 17: How many different items are sold
unique_items_sold = chipo['item_name'].nunique()
print(f"\nNumber of different items sold: {unique_items_sold}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the dataset
file_path = r"C:\Users\lovey\Downloads\us-marriages-divorces-1867-2014_Q6.csv"
data = pd.read_csv(file_path)

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per 1000', marker='o')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per 1000', marker='o')

# Label axes and title
plt.xlabel('Year', fontsize=14)
plt.ylabel('Per Capita Count', fontsize=14)
plt.title('U.S. Marriages and Divorces Per Capita (1867-2014)', fontsize=16)

# Add legend
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Import the dataset
file_path = r"C:\Users\lovey\Downloads\roman-emperor-reigns_Q9.csv"
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(file_path)

# Count the assassination statuses
counts = data['Cause_of_Death'].value_counts()
assassinated = counts.get('Assassinated', 0)
other = counts.sum() - assassinated

# Prepare data for the pie chart
labels = ['Assassinated', 'Other']
sizes = [assassinated, other]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['orange', 'skyblue'])
plt.title('Fraction of Roman Emperors Assassinated', fontsize=16)
plt.axis('equal')  # Ensures the pie chart is a circle
plt.show()
```
